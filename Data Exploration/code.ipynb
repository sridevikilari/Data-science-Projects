```python
#Define filter and apply it to calculate the mean for related automobiles which have price > 0
volvo_filter1 = ((automobile['make'] == "volvo") & (automobile['fuel-type'] == "diesel") & (automobile['price'] > 0))
volvo_mean1 = automobile.loc[volvo_filter1,'price'].mean()
volvo_mean1
```

```python
#To identify NaN values in data
automobile.isnull().sum(axis = 0)
```

```python
#Filling all NaN values with mean of respective column
automobile['normalized-losses'] = automobile['normalized-losses'].fillna(automobile['normalized-losses'].mean(axis=0)).round(2)
automobile['bore']              = automobile['bore'].fillna(automobile['bore'].mean(axis=0))
automobile['stroke']            = automobile['stroke'].fillna(automobile['stroke'].mean(axis=0))
automobile['horsepower']        = automobile['horsepower'].fillna(automobile['horsepower'].mean(axis=0))
automobile['peak-rpm']          = automobile['peak-rpm'].fillna(automobile['peak-rpm'].mean(axis=0))
automobile['price']             = automobile['price'].fillna(automobile['price'].mean(axis=0)).round(2)

#Verify the results
automobile.isnull().sum(axis = 0)
```

```python
# Task 2: Data Exploration
# 1. Chosen the columns 'body-style' for nominal values, 'symboling' for ordinal Values, and 'length' for numerical values
##importing matplot library
import matplotlib.pyplot as plt
```

```python
#Piegraph for car body-style (nominal values)
automobile['body-style'].value_counts().plot(kind='pie',autopct='%.2f%%', figsize = (6,6))
plt.title("Percentage of automobiles by body-style")
plt.ylabel("")
plt.show()   
```

```python
##barplot of symboling (ordinal values)
automobile['symboling'].value_counts().plot(
    kind    = 'bar',
    title   = 'Automobile Symboling Info',
    legend  = True,
    figsize = (7,5),
    color   = 'pink',fontsize=10
)
plt.ylabel('Number of automobiles')
plt.xlabel('Symboling');
plt.show()
```

```python
##histogram for car length (numerical values)
automobile['length'].plot(kind='hist',bins=20, figsize = (8,6))
plt.title('Histogram of Automobile Lengths')
plt.xlabel('length')
plt.show()
```

```python
#comparing city-mpg and fuel-type
automobile.boxplot(column='city-mpg',by='fuel-type')
plt.title("")
plt.suptitle("City-mpg distribution by Fuel-type")
plt.xlabel("fuel-type")
plt.ylabel("city-mpg")
plt.show()
```

```python
#comparing engine-size and price
automobile.plot(kind='scatter',x='engine-size',y='price',c='blue')
plt.title("Scatter plot of Engine-size vs Price")
plt.show()
```

```python
#comparing horsepower and highway-mpg
automobile.plot(kind='hexbin', x="horsepower",y="highway-mpg",gridsize=10)
plt.title("Hexbin of Horsepower vs Highway-mpg")
plt.show()
```

```python
##scatter matrix for numeric columns
from pandas.plotting import scatter_matrix
scatter_matrix(automobile,figsize=(25,25),diagonal='hist')
plt.title("Scatter matrix")
plt.show()
```
